### Eyeballing dataset distributions with histograms

```python
address = 'C:/Users/Patron/Desktop/DS_ESS1/Data/mtcars.csv'

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

#make car names as the row indices 
cars.index = cars.car_names
mpg = cars['mpg']
mpg.plot( kind = 'hist', title = "MPG")

#histogram mpg as category and freq as y-values
#see 2:46
```

```python
#use plt function to make the same graph 3:10

plt.hist(mpg, facecolor = 'g')
plt.title("MPG")

plt.show()
```

```python
# seaborn library .displot(y) function 
sb.displot(mpg)
```

###  Seeing scatterplots in action

```python
# df plot func for scatter plot between hp and mpg (x and y)
#set c(color), s(marker size)
#4:43

cars.plot(kind = 'scatter', x= 'hp', y='mpg', c= ['salmon'], s = 100, marker = '+')
```

```python
#seaborn scatter plot
# .regplot(x=, y=, data=, scatter = True)
sb.regplot(x = cars['hp'], y = cars['mpg'], scatter = True)
```

```python
# alternative expression
sb.regplot(data = cars, x = 'hp', y='mpg', scatter = True)
```

### Generating a scatter plot matrix

```python
# seaborn: .pairplot(df obj)
sb.pairplot(cars)
plt.show()
```

```python
# make pairplot for a subset (mpg, disp, hp, wt)
# ? need plt.show()?
df_sub = cars.sample(n=3, axis = 1) #random select 3 columns (3 variables)
df_sub.shape
```

```python
sb.pairplot(df_sub)
plt.show()
```
