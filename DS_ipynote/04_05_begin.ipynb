### The simplest time series plot

```python
addr =  "./../Data/Superstore-Sales.csv"
df = pd.read_csv(addr, index_col = 'Order Date', encoding = 'cp1252', parse_dates=True)
#index col: use Order date to be the index of each row that contains other variables
df.head()
```

```python
df["Order Quantity"].plot() # plot order quantity as y values and the index- order date as x 
```

```python
df2 = df.sample(n=50, random_state = 25, axis = 0) 



df2['Order Quantity'].plot(grid = True,xlabel = "Order Date", ylabel = "Order Quantity",
                           title = 'Superstore-Sales', backend = 'matplotlib')
```
