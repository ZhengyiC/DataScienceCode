### The basics

```python
addr = 'C:/Users/Patron/Desktop/DS_ESS1/Data/mtcars.csv'
cars = pd.read_csv(addr)
cars.columns= ['cars_names', 'mpg','cyl','disp','hp','drat',
              'wt','qsec','vs','am','gear','carb']
cars.index = cars['cars_names']
#index the rows by cars' names
cars.head()
```

```python
#count unique values 
carb = cars['carb']
carb.value_counts()
#first column is the possible values for the vari
#sec col is the counts 
```

```python
nm =cars['cars_names']
nm.value_counts() 
## all unique names 
```

```python
cars_cat = cars[['cyl', 'vs','am','carb', 'gear']]
cars_cat.head()
```

```python
gear_grp = cars_cat.groupby('gear')
#describe the data set based on the gear grouping
gear_grp.describe()
#output a 3*32 table 
    # 3 rows for the 3 groups of different gears
    # 8 descriptive statistics for each vari (count,mean ... )
    # 4 other variables 
```

### Transforming variables to categorical data type

```python
#create categorical series for a variable
# add to the df obj cars
cars["group"] = pd.Series(cars.gear, dtype= 'category')
```

```python
cars['group'].dtypes
```

```python
cars['group'].value_counts()
```
