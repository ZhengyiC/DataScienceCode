```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline 
#command line config to make sure the fig printed in line 
rcParams['figure.figsize'] = 5,4 #set fig size width and height in inch
```

### Defining axes, ticks, and grids

```python
x = range(0,10) #0 to 9
y = list([1,2,3,4,0,4,3,2,1,0])

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1]) #dimension: left start co-ord, bottom start co-ord, width, height

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1]) #dimension: left start co-ord, bottom start co-ord, width, height

#set xy limit 

ax.set_xlim([0,9])
ax.set_ylim([0,5])

#set ticks 
ax.set_xticks([0,2,4,5,6,9])
ax.set_yticks([0,1,4,5])

print(f" left and right x lim: {ax.get_xlim()}") 
# lim follows the largest param (xticks' max > xlim's max --> follow xticks)

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1]) #dimension: left start co-ord, bottom start co-ord, width, height
ax.set_xlim([1,8])
ax.set_ylim([0,5])

ax.grid() # add grid
ax.plot(x,y)
```

### Generating multiple plots in one figure with subplots 

```python
fig = plt.figure()
fig,(ax1,ax2) = plt.subplots(1,2) #1 row and 2 col

ax1.set_xlim([0,8])
ax2.set_xlim([0,8])
ax1.plot(x)
ax2.plot(x,y)
```
