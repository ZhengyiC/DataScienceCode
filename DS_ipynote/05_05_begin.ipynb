### The Pearson Correlation

#### pearson corr coeff: 
    between 1 and -1
    use to uncover linear relation between variables 
    but cannot role out non-linear correlations
    can be used in machine learning algorithms to select independent variables 
#### data requirement to use pearson test
    normally distributed (checked the histogram)
    nearly related
    continuous, numerical variables 
        eg, am and gear values are more like categorical --> only fixed values 
    

```python
addr = 'C:/Users/Patron/Desktop/DS_ESS1/Data/mtcars.csv'
cars = pd.read_csv(addr)
cars.columns= ['cars_names', 'mpg','cyl','disp','hp','drat',
              'wt','qsec','vs','am','gear','carb']
```

```python
sb.pairplot(cars)
```

```python
x = cars[['mpg','hp', 'qsec', 'wt']]
sb.pairplot(x)
```

### Using scipy to calculate the Pearson correlation coefficient

```python
mpg = cars['mpg']
hp = cars['hp']
wt = cars['wt']

pearsonr_coeff, pval = pearsonr(mpg,hp)
print('pearson correlation coeff: %0.3f'%(pearsonr_coeff))
```

```python
pearsonr_coeff, pval = pearsonr(mpg,wt)
print('pearson correlation coeff: %0.3f'%(pearsonr_coeff))
```

```python
qseq = cars['qsec']

pearsonr_coef2, pval = pearsonr(mpg, qseq)
print('pearson correlation coeff: %0.3f'%(pearsonr_coef2))
```

### Using pandas to calculate the Pearson correlation coefficient

```python
corr = x.corr()
corr
```

### Using Seaborn to visualize the Pearson correlation coefficient

```python
sb.heatmap(corr, xticklabels = corr.columns.values, yticklabels = corr.columns.values)
```
